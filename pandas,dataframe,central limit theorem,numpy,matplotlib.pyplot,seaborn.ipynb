```python
#central limit theorem
sampled_means=[]
#for 1000 times
for i in range(0,1000):
    #take a random samples of 100 rows from the population,take the mean of the rows ,append to ampled_means
    sampled_means.append(house['SalePrice'].sample(n=100).mean())
#plot a histogram of sampled means
#it is clearly normally distributted and centered around 5000
pd.Series(sampled_means).hist(bins=100)
```
