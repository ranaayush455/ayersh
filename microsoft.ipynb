```python
from sklearn.model_selection import train_test_split #importing train and test for model
y = microsoft['HasDetections']
microsoft.drop(columns='HasDetections',inplace=True)

X_train,X_test,y_train,y_test = train_test_split( microsoft,y,test_size=0.2,random_state=24)
```

```python
object_columns = []
for column in microsoft.columns:
    if microsoft[column].dtype=='object':
        object_columns.append(column)
    else:
        pass
object_columns
```

```python
microsoft.dtypes
```

```python
#label encoding the categorical columns present in our train data
from sklearn.preprocessing import LabelEncoder

X_train,X_test,y_train,y_test = train_test_split(microsoft,y,test_size=0.2,random_state=24)

le=LabelEncoder()

for col in object_columns:
    X_train[col]=le.fit_transform(X_train[col].astype(str))
    X_test[col]=X_test[col].map(lambda s:'<unknown>' if s not in le.classes_ else s)
    le.classes_=np.append(le.classes_,'<unknown>')
    X_test[col]=le.transform(X_test[col])
    
X_test    
```

```python
from sklearn.decomposition import PCA
pca = PCA().fit(X_train)

plt.plot(np.cumsum(pca.explained_variance_ratio_))
#plt.plot(pca.explained_variance_ratio_)
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')   

pca.components_
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_skied=scaler.transform(X_test)
#MinMaxScaler(copy=True, feature_range=(0, 1))

X_train_scaled
```

```python
pca_1=PCA(n_components=10)
principalcomponents=pca_1.fit_transform(X_train_scaled)
df_train=pd.DataFrame(data=principalcomponents,columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10'])

df_train
                    
```

```python
principalcomponents=pca_1.transform(X_test_skied)
df_test=pd.DataFrame(data=principalcomponents,columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10'])

df_test
                    
```

```python
from sklearn.tree import DecisionTreeClassifier # importing decisiontreeclassifier from sklearn.tree to do classification

decision = DecisionTreeClassifier()

decision.fit(df_train,y_train)
predict = decision.predict(df_test)
predict
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=75,random_state=42)

```

```python
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
tuning={ 'n_estimators':np.arange(10,200,10),
          'max_depth':[5,6,7,8,9],
          'min_samples_leaf':[2,3,4,5,6],
          'min_samples_split':[6,7,8,9,10],
          'max_leaf_nodes':[2,3,4,5,6]}
         
rd = RandomizedSearchCV(estimator=rf,param_distributions=tuning,n_iter=150,cv=5,verbose=10,random_state=24)
rd.fit(df_train,y_train)
rd.best_params_
```

```python
from sklearn.model_selection import GridSearchCV
dictionary = {'n_estimators':np.arange(75,80,85),
              'max_depth':[2,4,6,7],
              'min_samples_split':[2,5,7,6],
              'min_samples_leaf':[2,4,5,6,7]
             
             }
grid = GridSearchCV(estimator=rf,param_grid=dictionary,cv=5)
grid.fit(df_train,y_train)
grid.best_params_
#grid.best_score_
```

```python
from sklearn.naive_bayes import GaussianNB
classify = GaussianNB()
classify.fit(df_train,y_train)
predic=classify.predict(df_test)
predic
```

```python
from sklearn.svm import SVC
classifier = SVC(C=10,kernel='linear',degree=4,random_state=42,probability=True)
classifier.fit(df_train,y_train)
prediction = classifier.predict(df_test)
prediction
```

```python
from sklearn.metrics import confusion_matrix,f1_score,balanced_accuracy_score,accuracy_score,roc_auc_score,roc_curve,precision_score,precision_recall_curve,roc_auc_score
import matplotlib.pyplot as plt
#confusion_matrix(y_test,pred)
confusion_matrix(y_test,prediction)
#importing confusin_matrix for for model performance and best model is here build by svm
y_pred_proba=classifier.predict_proba(df_test)[::,1]

y_pred_proba

fpr,tpr,thresholds=roc_curve(y_test,y_pred_proba)



plt.plot(fpr,tpr)

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

```
