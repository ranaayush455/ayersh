```python
import numpy as np
import pandas as pd
data_df = pd.read_csv('Amazon_reviews.csv')
data_df
```

```python
data_df['Label'] = data_df['Label'].map({'__label__1 ':0,'__label__2 ':1 })
data_df
```

```python
y = data_df['Label']
data_df.drop(columns='Label',axis=1,inplace=True)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data_df,y, random_state=44,test_size=0.2)
X_train
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.tokenize import RegexpTokenizer

tokenizer= RegexpTokenizer(r'\w+')
lemmatizer=WordNetLemmatizer()
stemmer=PorterStemmer()

def preprocessing(review):
    final_tokens =' '
    tokens= tokenizer.tokenize(review)
    pure_tokens=[token.lower() for token in tokens if token.lower() not in stopwords.words('english')]
    lemmas_tokens=[lemmatizer.lemmatize(pure_token) for pure_token in pure_tokens]
    
    final_tokens=final_tokens.join(lemmas_tokens)
    return final_tokens

X_train['Cleansed_text']=X_train['Review'].apply(preprocessing)

X_train['Cleansed_text']
```

```python
X_test['Cleansed_text']=X_test['Review'].apply(preprocessing)
X_test['Cleansed_text']
```
