```python
import pandas as pd
diabetes = pd.read_csv('diabetes.csv')#reading dataframe
diabetes
```

```python
diabetes.isnull().sum()#checking empty coloumns
```

```python
from sklearn.model_selection import train_test_split #importing train and test for model
y = diabetes['Outcome']
diabetes.drop(columns='Outcome',inplace=True)

X_train,X_test,y_train,y_test = train_test_split(diabetes,y,test_size=0.2,random_state=24)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier # importing decisiontreeclassifier from sklearn.tree to do classification

decision = DecisionTreeClassifier()

decision.fit(X_train,y_train)
predicted = decision.predict(X_test)
predicted
```

```python
comparison_df = pd.DataFrame({'actual':y_test,'predicted':predicted})#comparision of actual ytest and predicted 
comparison_df
```

```python
from sklearn.metrics import confusion_matrix,average_precision_score,f1_score,recall_score,roc_auc_score,precision_score,roc_curve
confusion_matrix(y_test,predicted)
```

```python
average_precision_score(y_test,predicted)
```

```python
f1_score(y_test,predicted)
```

```python
recall_score(y_test,predicted)
```

```python
roc_auc_score(y_test,predicted)
```

```python
precision_score(y_test,predicted)
```

```python
tpr,fpr,threshold= roc_curve(y_test,predicted)
```

```python
import matplotlib.pyplot as plt
#importing matplot library for 2d plotting
tpr,fpr,threshold = roc_curve(y_test,predicted)
plt.plot(tpr,fpr)# making roccurve
```

```python
y_pred_proba = decision.predict_proba(X_test)[::,1]  #to take probability estimates of positive class array we use ::1
y_pred_proba
```

```python
comparison_df = pd.DataFrame({'actual':y_test,'predicted':y_pred_proba})
comparison_df
```

```python
precision_score(y_test,y_pred_proba)# precision score
```

```python
import matplotlib.pyplot as plt
auc = roc_auc_score(y_test,y_pred_proba)
plt.plot(tpr,fpr,label='auc'+str(auc))
plt.xlabel('false positive rate')
plt.ylabel('true negative rate')
plt.title('AUC_ROC_curve')
plt.legend(loc=8)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score
from sklearn.ensemble import AdaBoostClassifier

adaboost=AdaBoostClassifier(algorithm='SAMME')
learning_rate= list(np.arange(0.1,1,0.1))
f1_scores=[]

for lr in learning_rate:
    adaboost=AdaBoostClassifier(algorithm='SAMME',learning_rate=lr,n_estimators=100)
    adaboost.fit(X_train,y_train)
    y_ada_pred=adaboost.predict(X_test)
    score=f1_score(y_test,y_ada_pred)
    f1_scores.append(score)
plt.plot(learning_rate,f1_scores)
    
                    
```
