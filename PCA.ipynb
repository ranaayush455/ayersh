```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
diabetes=pd.read_csv('diabetes.csv')
diabetes
```

```python
from sklearn.model_selection import train_test_split #importing train and test for model
y = diabetes['Outcome']
diabetes.drop(columns='Outcome',inplace=True)

X_train,X_test,y_train,y_test = train_test_split( diabetes,y,test_size=0.2,random_state=24)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
pca = PCA().fit(X_train)

plt.plot(np.cumsum(pca.explained_variance_ratio_))
#plt.plot(pca.explained_variance_ratio_)
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')   

pca.components_
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_skied=scaler.transform(X_test)
#MinMaxScaler(copy=True, feature_range=(0, 1))

X_train_scaled
```

```python
pca_1=PCA(n_components=5)
principalcomponents=pca_1.fit_transform(X_train_scaled)
df_train=pd.DataFrame(data=principalcomponents,columns=['PC1','PC2','PC3','PC4','PC5'])

df_train
                    
```

```python

principalcomponents=pca_1.transform(X_test_skied)
df_test=pd.DataFrame(data=principalcomponents,columns=['PC1','PC2','PC3','PC4','PC5'])

df_test
                    
```

```python
from sklearn.tree import DecisionTreeClassifier # importing decisiontreeclassifier from sklearn.tree to do classification

decision = DecisionTreeClassifier()

decision.fit(df_train,y_train)
predict = decision.predict(df_test)
predict
```

```python
from sklearn.naive_bayes import GaussianNB
classify = GaussianNB()
classify.fit(df_train,y_train)
predicted=classify.predict(df_test)
predicted
```

```python
from sklearn.metrics import confusion_matrix,f1_score,balanced_accuracy_score,accuracy_score,roc_auc_score,roc_curve,precision_score,precision_recall_curve,roc_auc_score
import matplotlib.pyplot as plt
confusion_matrix(y_test,predict)
```

```python
precision_score(y_test,predict)
```
